## Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.


The difference between descriptive statistics and inferential statistics lies in what they do with data and the purpose they serve in data analysis.

### **1. Descriptive Statistics**

**Definition:**

Descriptive statistics are methods used to summarize, organize, and describe the important features of a dataset.
They describe what the data shows — they do not make predictions or generalizations beyond the data you have.

**Purpose:**

To provide a clear and simple summary of the sample data.

**Common Tools / Measures:**

- Measures of Central Tendency: Mean, Median, Mode

- Measures of Dispersion: Range, Variance, Standard Deviation

- Data Presentation: Tables, Graphs, Charts (e.g., bar charts, histograms)

**Example:**

Suppose you collect the exam scores of 50 students in a class.
You calculate:

- Mean = 72

- Median = 74

- Standard Deviation = 8

These values describe the performance of those 50 students.
You are not making conclusions about all students in the school — just describing the group you studied.

### **2. Inferential Statistics**

**Definition:**

Inferential statistics are methods used to make predictions, inferences, or generalizations about a population based on a sample of data.
They help us go beyond the data we have.

**Purpose:**

To draw conclusions or make decisions about a population using data from a sample.

**Common Tools / Techniques:**

- Hypothesis Testing (e.g., t-test, chi-square test)

- Confidence Intervals

- Regression Analysis

- ANOVA (Analysis of Variance)

**Example:**

You collect exam scores from 50 students (a sample) and want to know if the average score of all students in the school (the population) is greater than 70.
You perform a t-test and find that, with 95% confidence, the population mean score is between 68 and 74.

Here, you are using inferential statistics to infer about the entire school’s students from your sample data.

## Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.


### **What is Sampling in Statistics?**

**Definition:**
Sampling is the process of selecting a subset (sample) from a larger group (population) in order to collect data and make inferences about the whole population.

Because studying every individual in a population can be time-consuming, expensive, or impractical, statisticians use a sample to represent the population.

**Example of Sampling**

Suppose you want to know the average height of students in a university with 10,000 students.
Instead of measuring everyone, you might select 200 students (a sample) and use their data to estimate the average height of all 10,000 students (the population).

**Types of Sampling Methods**

There are many sampling methods, but two common ones are random sampling and stratified sampling.

### **1. Random Sampling**

**Definition:**

In random sampling, every individual in the population has an equal chance of being selected.

**Purpose:**

To reduce bias and ensure that the sample represents the population fairly.

**How it works:**

- Assign a number to each individual in the population.

- Use a random method (e.g., random number generator or lottery) to choose participants.

**Example:**

If there are 1,000 students, and you need a sample of 100, you randomly pick 100 names.
Each student has the same probability of being selected.

**Advantages:**

- Simple and unbiased.

- Easy to understand and implement.

**Disadvantages:**

May not represent all subgroups accurately (for example, gender or age differences may be unevenly represented by chance).

### **2. Stratified Sampling**

**Definition:**

In stratified sampling, the population is divided into subgroups (strata) that share similar characteristics (e.g., gender, age, grade level), and random samples are taken from each stratum.

**Purpose:**

To ensure that each subgroup is represented proportionally in the final sample.

**How it works:**

1. Divide the population into strata (e.g., male/female, departments, age groups).

2. Randomly select samples within each stratum based on its size in the population.

**Example:**

In a university with:

- 60% female students

- 40% male students

If you need a sample of 100 students, you select:

- 60 females randomly from all females

- 40 males randomly from all males

This ensures both genders are fairly represented.

**Advantages:**

- More accurate and representative.

- Reduces sampling bias.

**Disadvantages:**

- More complex to organize and execute.

- Requires prior knowledge of population structure.

### **Summary Table**

```python
| Feature            | Random Sampling                                   | Stratified Sampling                                                          |
| ------------------ | ------------------------------------------------- | ---------------------------------------------------------------------------- |
| **Definition**     | Every individual has an equal chance of selection | Population is divided into strata, and random samples are taken from each    |
| **Representation** | May not represent all subgroups                   | Ensures representation of all key subgroups                                  |
| **Complexity**     | Simple to carry out                               | More complex; requires subgroup information                                  |
| **Example**        | Randomly selecting 100 students from 1,000        | Selecting 60 females and 40 males from 1,000 students (based on proportions) |
```

## Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.


### **Measures of Central Tendency: Mean, Median, and Mode**

**Definition:**

Measures of central tendency are statistical values that describe the center or typical value of a dataset.
They help summarize a large amount of data with a single representative value.

The three most common measures are mean, median, and mode.

Screenshot (10).png

### **2. Median**

**Definition:**

The median is the middle value when all data are arranged in order (ascending or descending).
If there’s an even number of observations, the median is the average of the two middle values.

**Example:**

Data: 10, 20, 30, 40, 50
→ Median = 30 (the middle value)

If the data are 10, 20, 30, 40:
→ Median = (20 + 30) / 2 = 25

**Key Point:**

- Not affected by outliers.

- Useful when data are skewed (not symmetrical).

### **3. Mode**

**Definition:**

The mode is the value that occurs most frequently in a dataset.

**Example:**

Data: 5, 7, 7, 9, 10, 10, 10
→ Mode = 10 (it appears most often)

**Key Point:**

- There can be one mode (unimodal), two modes (bimodal), or many modes (multimodal).

- Useful for categorical data (e.g., most common color, brand, or choice).

### **Importance of Measures of Central Tendency**

```python
| **Reason**                             | **Explanation**                                                                                                                |
| -------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------ |
| **1. Simplifies large data**           | They summarize a large dataset into one representative number, making it easier to understand.                                 |
| **2. Basis for comparison**            | Different groups or time periods can be compared using a single average value.                                                 |
| **3. Helps identify trends**           | Mean, median, and mode help detect patterns or shifts in data over time.                                                       |
| **4. Useful in decision-making**       | Businesses, researchers, and policymakers use these measures to make informed decisions (e.g., average income, average score). |
| **5. Foundation for further analysis** | Many statistical techniques (like variance and standard deviation) are based on these central measures.                        |
```

## Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?


### **Skewness and Kurtosis in Statistics**

Skewness and kurtosis are measures of the shape of a data distribution.
While measures like the mean, median, and mode describe the center of data, skewness and kurtosis describe how data are spread and shaped around that center.

### **1. Skewness**

**Definition:**

Skewness measures the asymmetry (lack of symmetry) of a distribution.

- A symmetrical distribution (like a perfect bell curve) has a skewness = 0.

- A skewed distribution is one where one tail is longer or fatter than the other.

**Types of Skewness:**

**➤ a. Positive Skew (Right Skewed)**

- The tail on the right side of the distribution is longer.

- Most data values are concentrated on the left, with a few large values stretching the tail to the right.

- Mean > Median > Mode

**Example:**

Income distribution — most people earn moderate amounts, but a few earn extremely high salaries.

**Interpretation:**

A positive skew implies that the dataset has a few unusually high values (outliers) pulling the mean to the right.

**➤ b. Negative Skew (Left Skewed)**

- The tail on the left side is longer.

- Most data values are concentrated on the right, with a few small values stretching the tail to the left.

- Mean < Median < Mode

**Example:**

Age at retirement — most people retire around 60–65, but a few retire very early.

### **Summary of Skewness**

```python
| Type              | Shape             | Relationship         | Example             |
| ----------------- | ----------------- | -------------------- | ------------------- |
| **Positive Skew** | Tail to the right | Mean > Median > Mode | Income distribution |
| **Negative Skew** | Tail to the left  | Mean < Median < Mode | Age at retirement   |
| **Zero Skew**     | Symmetrical       | Mean = Median = Mode | Normal distribution |
```

### **2. Kurtosis**

**Definition:**

Kurtosis measures the "peakedness" or tailedness of a distribution — that is, how heavy or light the tails are compared to a normal distribution.

**Types of Kurtosis:**

**➤ a. Mesokurtic (Normal Kurtosis, = 3)**

- This is the normal distribution.

- Moderate peak and tails.

- Example: Standard normal curve.

**➤ b. Leptokurtic (High Kurtosis, > 3)**

- Sharper peak and heavier tails than a normal distribution.

- Indicates more data in the tails (more outliers).

**Example:**

Stock market returns (many values near the mean but some extreme highs/lows).

**➤ c. Platykurtic (Low Kurtosis, < 3)**

- Flatter peak and thinner tails than a normal distribution.

- Indicates fewer extreme values.

**Example:**

Heights of adults — most values are close to the mean, few extremes.

### **Summary Table**

```python
| Measure      | Description        | Indicates                      |
| ------------ | ------------------ | ------------------------------ |
| **Skewness** | Asymmetry of data  | Direction of tail (left/right) |
| **Kurtosis** | Peakedness of data | Thickness of tails (outliers)  |
```

## Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28] (Include your Python code and output in the code box below.)


```python
# Program to compute mean, median, and mode

# Import statistics module
import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean, median, and mode
mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)

# Display the results
print("Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

**Output:**

```python
Numbers: [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
Mean: 19.6
Median: 19
Mode: 12
```

## Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] (Include your Python code and output in the code box below.)

```python
# Program to compute covariance and correlation coefficient

import numpy as np

# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate covariance matrix
cov_matrix = np.cov(list_x, list_y, bias=False)  # bias=False → sample covariance

# Extract covariance value
covariance = cov_matrix[0, 1]

# Calculate correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0, 1]

# Display results
print("List X:", list_x)
print("List Y:", list_y)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

**Output:**

```python
List X: [10, 20, 30, 40, 50]
List Y: [15, 25, 35, 45, 60]
Covariance: 225.0
Correlation Coefficient: 0.989743318610787
```

## Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] (Include your Python code and output in the code box below.)


```python
# Program to draw a boxplot and identify outliers

import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of Given Data")
plt.xlabel("Values")
plt.show()

# Calculate quartiles and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Display results
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Q3 - Q1): {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

**Output:**

```python
Q1 (25th percentile): 18.75
Q3 (75th percentile): 23.75
IQR (Q3 - Q1): 5.0
Lower Bound: 11.25
Upper Bound: 31.25
Outliers: [35]
```

## Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. ● Explain how you would use covariance and correlation to explore this relationship. ● Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000] (Include your Python code and output in the code box below.)


```python
# Program to compute covariance and correlation between advertising spend and daily sales

import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Calculate covariance matrix
cov_matrix = np.cov(advertising_spend, daily_sales, bias=False)
covariance = cov_matrix[0, 1]

# Calculate correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

# Display results
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

**Output:**

```python
Advertising Spend: [200, 250, 300, 400, 500]
Daily Sales: [2200, 2450, 2750, 3200, 4000]
Covariance: 77500.0
Correlation Coefficient: 0.991893
```

## Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product. ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use. ● Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7] (Include your Python code and output in the code box below.)


```python
# Program to summarize and visualize customer satisfaction survey data

import matplotlib.pyplot as plt
import numpy as np
import statistics

# Given survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary statistics
mean_score = statistics.mean(survey_scores)
median_score = statistics.median(survey_scores)
mode_score = statistics.mode(survey_scores)
std_dev = np.std(survey_scores, ddof=1)  # sample standard deviation

# Display summary statistics
print("Customer Satisfaction Summary:")
print(f"Mean: {mean_score:.2f}")
print(f"Median: {median_score}")
print(f"Mode: {mode_score}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Minimum: {min(survey_scores)}")
print(f"Maximum: {max(survey_scores)}")
print(f"Number of Responses: {len(survey_scores)}")

# Plot histogram
plt.hist(survey_scores, bins=range(1, 12), color="skyblue", edgecolor="black")
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Number of Customers")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Output**

```python
Customer Satisfaction Summary:
Mean: 7.40
Median: 7
Mode: 7
Standard Deviation: 1.67
Minimum: 4
Maximum: 10
Number of Responses: 15
```
