## Question 1: What is a random variable in probability theory?
In probability theory, a random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment.

### **More precisely:**

- A random variable maps outcomes to numbers.

For example, when you roll a die, the outcome is one of {1, 2, 3, 4, 5, 6}. You can define a random variable X that assigns to each outcome its face value.

- It allows us to use mathematical tools (like algebra and calculus) to analyze randomness.

- Despite the name, a random variable is not actually random nor a variable‚Äîit‚Äôs a deterministic function applied to a random outcome.

### **Types of random variables:**

- **Discrete:** Takes on countable values (e.g., number of heads in 10 coin flips).

- **Continuous:** Takes values in a continuum (e.g., the time it takes for a bus to arrive).

### **Why random variables matter:**

They let us define important concepts like probability distributions, expectation, variance, and covariance, which describe and quantify randomness.

## Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution. (Include your Python code and output in the code box below.)


Screenshot (37).png

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a Normal(50, 5)
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean_val = np.mean(data)
std_val = np.std(data)

mean_val, std_val

# Plot histogram
plt.hist(data, bins=30)
plt.title("Histogram of Normally Distributed Data (mean=50, sd=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Computed Mean:** ~50.05

**Computed Standard Deviation:** ~4.91

The histogram plotted above visualizes the distribution.

## Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval. ‚óè Write the Python code to compute the mean sales and its confidence interval. (Include your Python code and output in the code box below.)

## **‚úÖ Python Code (Mean and 95% Confidence Interval)**

```python
import numpy as np
from math import sqrt

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.array(daily_sales)

mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)   # sample std
n = len(data)

# 95% CI using CLT (z = 1.96)
z = 1.96
margin_error = z * std_sales / sqrt(n)
ci_lower = mean_sales - margin_error
ci_upper = mean_sales + margin_error

mean_sales, (ci_lower, ci_upper)
```

### **‚úÖ Output Summary**

- **Mean daily sales:** 248.25

- **95% Confidence Interval:** (240.68, 255.82)

This means we are 95% confident that the true average daily sales lie **between about 241 and 256 units.**
